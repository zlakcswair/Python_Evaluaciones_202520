```python
# Parte 1: Clase Base Robot
class Robot:
    def __init__(self, name, model):
        self.name = name
        self.model = model
        self.battery = 100.0
        self.distance_traveled = 0.0
        self.speed = 0.0
        self.direction = 0
        print(f"Robot {name} modelo {model} creado")

    def __str__(self):
        return f"{self.name} ({self.model}) - Batería: {self.battery:.1f}% - Distancia: {self.distance_traveled:.2f}km"

    def move(self, speed, direction, time_minutes):
        print(f"\n{self.name} intentando moverse: {speed}km/h, {direction}°, {time_minutes}min")
        print(f"Batería antes: {self.battery:.1f}%")
        print(f"Distancia antes: {self.distance_traveled:.2f}km")

        self.speed = speed
        self.direction = direction % 360

        distancia = speed * (time_minutes / 60)
        consumo_bateria = distancia * 10

        print(f"Distancia planeada: {distancia:.2f}km")
        print(f"Consumo de batería estimado: {consumo_bateria:.1f}%")

        if consumo_bateria > self.battery:
            distancia_real = self.battery / 10
            self.distance_traveled += distancia_real
            print(f"¡Batería insuficiente! Solo puede recorrer {distancia_real:.2f}km")
            self.battery = 0
            print(f"¡{self.name} se quedó sin batería!")
        else:
            self.distance_traveled += distancia
            self.battery -= consumo_bateria
            print(f"Movimiento completado: {distancia:.2f}km recorridos")
            if self.battery < 20:
                print(f"¡{self.name}: Batería baja!")

        print(f"Batería después: {self.battery:.1f}%")
        print(f"Distancia después: {self.distance_traveled:.2f}km")

    def charge(self, time_minutes):
        print(f"\n{self.name} cargando por {time_minutes} minutos")
        print(f"Batería antes: {self.battery:.1f}%")

        carga = time_minutes * 2
        self.battery += carga

        if self.battery > 100:
            print(f"Carga excedente: {self.battery - 100:.1f}% (se descarta)")
            self.battery = 100

        print(f"Batería después: {self.battery:.1f}%")

    def get_status(self):
        print(f"\n--- Estado completo de {self.name} ---")
        print(f"Modelo: {self.model}")
        print(f"Batería: {self.battery:.1f}%")
        print(f"Distancia recorrida: {self.distance_traveled:.2f}km")
        print(f"Velocidad actual: {self.speed}km/h")
        print(f"Dirección: {self.direction}°")
# ----------------------------------------------------------------------------------------------------
# Parte 2: Herencia - Tipos de Robot
class WorkerRobot(Robot):
    def __init__(self, name, model):
        super().__init__(name, model)
        self.cargo = 0.0

    def move(self, speed, direction, time_minutes):
        print(f"\n{self.name} (con {self.cargo}kg de carga) intentando moverse")
        print(f"Batería antes: {self.battery:.1f}%")

        consumo_extra = 0
        if self.cargo > 0:
            distancia = speed * (time_minutes / 60)
            consumo_extra = distancia * 2
            print(f"Consumo adicional por carga: {consumo_extra:.1f}%")

        super().move(speed, direction, time_minutes)

        if self.battery > 0 and consumo_extra > 0:
            if consumo_extra > self.battery:
                self.battery = 0
                print(f"¡{self.name} se quedó sin batería por la carga extra!")
            else:
                self.battery -= consumo_extra
                print(f"Consumo adicional aplicado: -{consumo_extra:.1f}%")
                print(f"Batería final: {self.battery:.1f}%")

    def load_cargo(self, weight):
        print(f"\n{self.name} cargando {weight}kg")
        print(f"Carga antes: {self.cargo}kg")
        self.cargo += weight
        print(f"Carga después: {self.cargo}kg")

    def unload_cargo(self):
        print(f"\n{self.name} descargando {self.cargo}kg")
        self.cargo = 0
        print("Carga descargada completamente")

    def get_status(self):
        super().get_status()
        print(f"Carga actual: {self.cargo}kg")

class GuardRobot(Robot):
    def __init__(self, name, model):
        super().__init__(name, model)
        self.alert_mode = False

    def charge(self, time_minutes):
        print(f"\n{self.name} cargando (modo alerta: {'ON' if self.alert_mode else 'OFF'})")
        print(f"Batería antes: {self.battery:.1f}%")

        if self.alert_mode:
            tiempo_efectivo = time_minutes * 0.5
            print(f"Carga al 50% por modo alerta: {time_minutes}min → {tiempo_efectivo}min efectivos")
        else:
            tiempo_efectivo = time_minutes
            print(f"Carga normal: {tiempo_efectivo}min efectivos")

        super().charge(tiempo_efectivo)

    def toggle_alert(self):
        print(f"\n{self.name} cambiando modo alerta")
        print(f"Modo alerta antes: {'ON' if self.alert_mode else 'OFF'}")
        self.alert_mode = not self.alert_mode
        print(f"Modo alerta después: {'ON' if self.alert_mode else 'OFF'}")

    def get_status(self):
        super().get_status()
        print(f"Modo alerta: {'Activado' if self.alert_mode else 'Desactivado'}")

# ----------------------------------------------------------------------------------------------------
# Parte 3: Composición - Clase RobotTeam
class RobotTeam:
    def __init__(self, team_name):
        self.team_name = team_name
        self.robots = []
        self.missions_completed = 0
        print(f"Equipo {team_name} formado")

    def add_robot(self, robot):
        print(f"\n{robot.name} solicitando unirse al equipo {self.team_name}")
        print(f"Robots en equipo antes: {len(self.robots)}")
        self.robots.append(robot)
        print(f"Robots en equipo después: {len(self.robots)}")

    def __len__(self):
        return len(self.robots)

    def charge_all(self, time_minutes):
        print(f"\nCargando todos los robots del equipo {self.team_name}...")
        print(f"Tiempo de carga: {time_minutes} minutos")
        for i, robot in enumerate(self.robots, 1):
            print(f"\n{i}. Cargando {robot.name}")
            robot.charge(time_minutes)

    def team_status(self):
        print(f"\n===== Estado del Equipo: {self.team_name} =====")
        print(f"Total de robots: {len(self)}")
        print(f"Misiones completadas: {self.missions_completed}")

        for i, robot in enumerate(self.robots, 1):
            print(f"\n{i}. {robot}")
            robot.get_status()

    def complete_mission(self):
        print(f"\nEquipo {self.team_name} completando misión")
        print(f"Misiones antes: {self.missions_completed}")
        self.missions_completed += 1
        print(f"Misiones después: {self.missions_completed}")

# ----------------------------------------------------------------------------------------------------
# Parte 4: Código de Prueba
# DEMOSTRACIÓN COMPLETA DEL SISTEMA
print("=" * 60)
print("DEMOSTRACIÓN DEL SISTEMA DE ROBOTS")
print("=" * 60)

# Creación de robots
print("\n" + "=" * 30)
print("FASE 1: CREACIÓN DE ROBOTS")
print("=" * 30)

robot_basico = Robot("R2D2", "Astromech")
robot_basico2 = Robot("C3PO", "Protocolo")
robot_trabajador = WorkerRobot("Wall-E", "Compactador")
robot_guardian = GuardRobot("T-800", "Terminator")

# Prueba de movimiento CON batería suficiente (caso positivo)
print("\n" + "=" * 30)
print("FASE 2: MOVIMIENTO CON BATERÍA SUFICIENTE (CASO POSITIVO)")
print("=" * 30)

robot_basico.move(5, 45, 30)  # 5 km/h, 45°, 30 minutos (solo 2.5 km)

# Prueba de movimiento que agota la batería (caso negativo)
print("\n" + "=" * 30)
print("FASE 3: MOVIMIENTO QUE AGOTA LA BATERÍA (CASO NEGATIVO)")
print("=" * 30)

robot_basico2.move(50, 90, 120)  # 50 km/h, 90°, 120 minutos (debería agotar batería)

# Prueba de WorkerRobot con carga y batería suficiente
print("\n" + "=" * 30)
print("FASE 4: ROBOT TRABAJADOR CON CARGA Y BATERÍA SUFICIENTE")
print("=" * 30)

# Primero cargamos el robot completamente
robot_trabajador.charge(50)  # Cargar al 100%
robot_trabajador.load_cargo(20)  # Cargar 20kg (no demasiado)
robot_trabajador.move(10, 180, 30)  # 10 km/h, 180°, 30 minutos (5 km)

# Prueba de WorkerRobot con carga excesiva que agota la batería
print("\n" + "=" * 30)
print("FASE 5: ROBOT TRABAJADOR CON CARGA EXCESIVA")
print("=" * 30)

robot_trabajador2 = WorkerRobot("Cargador", "Pesado")
robot_trabajador2.load_cargo(100)  # Carga muy pesada
robot_trabajador2.move(20, 0, 60)  # 20 km/h, 0°, 60 minutos

# Prueba de GuardRobot con modo alerta
print("\n" + "=" * 30)
print("FASE 6: ROBOT GUARDIÁN CON MODO ALERTA")
print("=" * 30)

robot_guardian.toggle_alert()
robot_guardian.move(15, 270, 30)  # 15 km/h, 270°, 30 minutos

# Pruebas de carga
print("\n" + "=" * 30)
print("FASE 7: PRUEBAS DE CARGA")
print("=" * 30)

robot_basico.charge(20)
robot_basico2.charge(50)
robot_trabajador.charge(30)
robot_trabajador2.charge(60)
robot_guardian.charge(40)  # Carga más lenta por modo alerta

# Creación y gestión de equipo
print("\n" + "=" * 30)
print("FASE 8: GESTIÓN DE EQUIPO")
print("=" * 30)

equipo = RobotTeam("Alpha Strike")
equipo.add_robot(robot_basico)
equipo.add_robot(robot_basico2)
equipo.add_robot(robot_trabajador)
equipo.add_robot(robot_trabajador2)
equipo.add_robot(robot_guardian)

equipo.charge_all(15)
equipo.complete_mission()
equipo.complete_mission()

# Estado final detallado
print("\n" + "=" * 30)
print("FASE 9: ESTADO FINAL DETALLADO")
print("=" * 30)

robot_basico.get_status()
robot_basico2.get_status()
robot_trabajador.get_status()
robot_trabajador2.get_status()
robot_guardian.get_status()

print("\n" + "=" * 30)
print("RESUMEN DEL EQUIPO")
print("=" * 30)

equipo.team_status()

print("\n" + "=" * 60)
print("DEMOSTRACIÓN COMPLETADA EXITOSAMENTE")
print("=" * 60)
```
